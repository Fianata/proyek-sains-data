# **Data Preprocessing**

## **1. Pendahuluan**

### **1.1 Pengertian Data Preprocessing**

**Data preprocessing** merupakan proses fundamental dan tahapan penting dalam data science yang bertujuan untuk mengubah data mentah (raw data) menjadi format yang lebih bersih, terstruktur, dan siap untuk digunakan dalam analisis lebih lanjut.

Dalam konteks machine learning, data preprocessing adalah tahap kritis yang sangat menentukan tingkat keberhasilan suatu model. Data yang tidak diproses dengan baik dapat menyebabkan model belajar pola yang salah, menghasilkan prediksi yang tidak akurat, dan pada akhirnya gagal memberikan insight yang bermakna bagi pengambilan keputusan.

Proses ini mencakup berbagai teknik seperti handling missing values, normalisasi data, encoding variabel kategorikal, dan feature scaling yang akan kita bahas lebih detail pada bagian berikutnya.

## **1.2 Pentingnya Data Preprocessing dalam Machine Learning**

- **Peningkatan Kualitas Model**
Data yang telah melalui preprocessing menjadi lebih bersih dan terstruktur, sehingga memungkinkan algoritma machine learning untuk:
  - Mempelajari pola data dengan lebih akurat
  - Mengurangi noise yang dapat mengganggu proses pembelajaran
  - Meningkatkan generalisasi model pada data baru

- **Penanganan Masalah Data**
Proses preprocessing membantu mengidentifikasi dan menangani berbagai masalah data seperti:
  - **Anomali dan Outlier**: Mendeteksi nilai ekstrem yang dapat mempengaruhi hasil
  - **Missing Values**: Menangani data yang hilang dengan berbagai strategi
  - **Bias Data**: Mengurangi bias yang tidak diinginkan dalam dataset

-  **Optimasi Performa Komputasi**
   - **Reduksi Dimensi**: Mengurangi kompleksitas data tanpa kehilangan informasi penting
   - **Normalisasi**: Menyamakan skala data untuk mempercepat konvergensi algoritma
   - **Efisiensi Memori**: Mengoptimalkan penggunaan resources komputasi

- **Konsistensi dan Standardisasi**
  - **Format Konsisten**: Menyamakan format data dari berbagai sumber
  - **Skala Seragam**: Memastikan semua fitur berada dalam skala yang comparable
  - **Integrasi Data**: Memudahkan penggabungan data dari multiple sources

- **Peningkatan Interpretasi Hasil**
  - Data yang terstruktur dengan baik memudahkan interpretasi hasil model
  - Feature engineering melalui preprocessing dapat meningkatkan insight yang didapat
  - Visualisasi data menjadi lebih meaningful setelah preprocessing

- **Robustness Model**
  - Mengurangi overfitting dengan handling outliers dan noise
  - Meningkatkan stabilitas model terhadap variasi data input
  - Memastikan reliability model dalam production environment

> **Best Practice**: Preprocessing yang baik dapat meningkatkan akurasi model hingga 30% dan mengurangi waktu training secara signifikan.

# 2. Tahapan Data Preprocessing

## 2.1 Data Cleaning (Pembersihan Data)

###  Menangani Missing Values
**Penghapusan Data**
- Cocok untuk dataset besar dimana missing values hanya sebagian kecil
- Dapat dilakukan dengan `dropna()` pada pandas
- Risiko: Dapat mengurangi jumlah sampel secara signifikan jika missing values banyak

**Imputasi Statistik**
- Mengisi nilai yang hilang dengan mean/median (numerik) atau mode (kategorikal)
- Implementasi: `SimpleImputer` dari scikit-learn
- Kelebihan: Mudah diimplementasikan dan cepat

**Advanced Imputation**
- Menggunakan machine learning algorithms untuk memprediksi missing values
- Contoh: KNN Imputer, Iterative Imputer (MICE)
- Cocok untuk dataset dengan pola missing yang kompleks

###  Menangani Outliers
**Deteksi Outliers:**
- **Metode Statistik**: Z-score (±3 standard deviations), IQR (Q1 - 1.5*IQR, Q3 + 1.5*IQR)
- **Visualisasi**: Box plots (mengidentifikasi whiskers), scatter plots (pattern anomalies)
- **Machine Learning**: Isolation Forest (tree-based), DBSCAN clustering (density-based), ABOD, LOF

**Penanganan Outliers:**
- **Removal**: Untuk outliers yang jelas merupakan kesalahan pengukuran/data entry
- **Capping**: Winsorizing - membatasi nilai outliers pada percentil ke-1 dan ke-99
- **Transformation**: Log transformation, square root transformation untuk data skewed

---

## 2.2 Data Transformation

###  Normalization dan Standardization
**Min-Max Scaling**
- Rumus: \( X_{\text{norm}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}} \)
- Cocok untuk algorithms yang bergantung pada distance metrics seperti KNN

**Z-score Standardization**
- Rumus: \( X_{\text{std}} = \frac{X - \mu}{\sigma} \)
- Hasil: mean = 0, standard deviation = 1
- Cocok untuk algorithms yang mengasumsikan data terdistribusi normal

###  Encoding Categorical Data
**One-Hot Encoding**
- Mengubah categorical variables menjadi binary vectors (0/1)
- Gunakan `pd.get_dummies()` atau `OneHotEncoder` dari sklearn
- Hati-hati dengan curse of dimensionality untuk categorical features dengan banyak kategori

**Label Encoding**
- Mengassign numerical values kepada categories (misal: red=1, blue=2, green=3)
- Cocok untuk ordinal data atau tree-based algorithms

**Target Encoding**
- Menggunakan mean of target variable untuk setiap category
- Berguna untuk high-cardinality features
- Risiko overfitting - perlu menggunakan smoothing atau cross-validation

###  Feature Engineering
**Polynomial Features**
- Membuat interaction terms dan polynomial terms (X², X³, X₁X₂)
- Dapat menangkap nonlinear relationships
- Implementasi: `PolynomialFeatures` dari sklearn

**Binning**
- Mengubah continuous variables menjadi categorical bins (age groups, income brackets)
- Mengurangi noise dan overfitting

**Domain-Specific Features**
- Membuat features berdasarkan domain knowledge (BMI dari height/weight, weekday/weekend dari timestamp)
- Sering menjadi features yang paling predictive

---

## 2.3 Data Reduction

###  Dimensionality Reduction
**Principal Component Analysis (PCA)**
- Linear transformation yang mengurangi dimensionality sambil mempertahankan variance
- Sangat useful untuk multicollinearity dan visualization

**t-SNE dan UMAP**
- Nonlinear techniques untuk visualisasi high-dimensional data
- Menjaga struktur lokal dari data

**Feature Selection**
- **Filter Methods**: SelectKBest berdasarkan statistical tests
- **Wrapper Methods**: Recursive Feature Elimination (RFE)
- **Embedded Methods**: Lasso regularization, feature importance dari tree models

###  Dataset Reduction
**Sampling Techniques**
- Random sampling, Stratified sampling (mempertahankan class distribution)
- Berguna untuk large datasets dan rapid prototyping

**Aggregation**
- Mengaggregate data ke level yang lebih tinggi (harian ke mingguan, individual ke kelompok)
- Mengurangi noise dan computational complexity

---

## 3. Best Practices dan Tips Praktis

###  Workflow Management
**Reproducibility**
- Selalu dokumentasikan semua preprocessing steps dengan comments dan versioning
- Gunakan random seed untuk stochastic processes

**Pipeline Development**
- Bangun automated pipelines dengan `sklearn.pipeline` untuk konsistensi
- Memudahkan deployment dan maintenance

**Version Control**
- Gunakan Git untuk preprocessing scripts dan configuration files
- Track perubahan parameter dan hasilnya

###  Common Pitfalls
**Data Leakage**
- Hindari menggunakan information dari test set dalam preprocessing
- Selalu fit transform hanya pada training data, lalu transform test data

**Over-engineering**
- Jangan membuat features yang terlalu complex tanpa business justification
- Start simple, then iterate

**Ignoring Domain Context**
- Selalu konsultasi dengan domain experts untuk memahami data
- Context is king dalam interpretasi results

###  Validation Strategies
**Cross-validation**
- Implementasikan preprocessing dalam setiap cross-validation fold
- Prevent data leakage dan dapatkan robust performance estimates

**Monitoring**
- Terus monitor impact of preprocessing pada model performance
- A/B test different preprocessing strategies

##  Studi Kasus: Deteksi Outlier dengan ABOD

### Pengenalan Metode ABOD (Angle-Based Outlier Detection)
ABOD adalah algoritma deteksi outlier yang khusus efektif untuk **high-dimensional data**. Berbeda dengan metode berbasis jarak, ABOD menggunakan pendekatan berbasis sudut:

**Intuisi Utama:**
- Titik inlier: Tetangga tersebar di berbagai sudut → varians sudut tinggi
- Titik outlier: Titik data terkonsentrasi dalam arah sempit → varians sudut rendah

**Keunggulan ABOD:**
-  Efektif di dimensi tinggi (kurang terkena curse of dimensionality)
-  Tidak membutuhkan asumsi distribusi data
-  Deteksi outlier yang lebih robust dibanding metode tradisional

 ## **Analisis dengan Scikit-learn**

 Diakarenakan collab tidak bisa menggunakan `pycaret` maka saya menggunakan `scikit-learn`

```python

if dataset is not None:
    # Import necessary libraries
    from sklearn.neighbors import LocalOutlierFactor
    from sklearn.ensemble import IsolationForest
    from sklearn.cluster import KMeans
    import numpy as np
    import pandas as pd


    # --- PERSIAPAN DATA UNTUK MODEL ---
    print("Mempersiapkan data untuk analisis...")
    numeric_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
    X = dataset[numeric_features]

    # --- 1. DETEKSI OUTLIER MENGGUNAKAN K-NN (Local Outlier Factor) ---
    print("Mendeteksi outlier dengan K-NN (LOF)...")
    lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
    outlier_lof = lof.fit_predict(X)
    # Menambahkan hasil ke dataset: 1 jika outlier, 0 jika normal
    dataset['outlier_knn'] = np.where(outlier_lof == -1, 1, 0)

    # --- 2. DETEKSI OUTLIER MENGGUNAKAN ISOLATION FOREST ---
    print("Mendeteksi outlier dengan Isolation Forest...")
    iso_forest = IsolationForest(contamination=0.1, random_state=42)
    outlier_iso = iso_forest.fit_predict(X)
    # Menambahkan hasil ke dataset: 1 jika outlier, 0 jika normal
    dataset['outlier_abod_alt'] = np.where(outlier_iso == -1, 1, 0)

    # --- 3. CLUSTERING DENGAN K-MEANS ---
    print("Melakukan clustering dengan K-Means...")
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X)
    # Menambahkan hasil cluster ke dataset
    dataset['cluster'] = 'Cluster ' + pd.Series(cluster_labels).astype(str)

    # --- 4. MEMBUAT KOLOM RINGKASAN OUTLIER ---
    print("Membuat kolom ringkasan outlier...")
    conditions = [
        (dataset['outlier_knn'] == 1) & (dataset['outlier_abod_alt'] == 1),
        (dataset['outlier_knn'] == 1) & (dataset['outlier_abod_alt'] == 0),
        (dataset['outlier_knn'] == 0) & (dataset['outlier_abod_alt'] == 1),
        (dataset['outlier_knn'] == 0) & (dataset['outlier_abod_alt'] == 0)
    ]
    choices = ['Outlier by Both', 'Outlier by K-NN Only', 'Outlier by ABOD/IF Only', 'Normal']
    dataset['outlier_type'] = np.select(conditions, choices, default='Unknown')

    print("\n✅ Semua analisis dengan scikit-learn selesai.")
```

```python
# Import library tambahan untuk visualisasi
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# --- PERSIAPAN DATA UNTUK VISUALISASI ---
# Pastikan DataFrame 'dataset' sudah ada dari skrip sebelumnya

# 1. Buat kolom label outlier tunggal untuk pewarnaan (1 jika outlier, 0 jika normal)
# Ini berguna untuk membedakan secara visual antara data normal dan semua jenis outlier
dataset['is_outlier_any'] = dataset['outlier_type'].apply(lambda x: 0 if x == 'Normal' else 1)

# 2. Siapkan data numerik dan lakukan scaling untuk PCA
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = dataset[numeric_columns]
X_scaled = StandardScaler().fit_transform(X)


# --- MEMBUAT GRID VISUALISASI 2x3 ---
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('Dashboard Analisis Dataset Iris', fontsize=20, fontweight='bold')

# 1. Komposisi Kategori Outlier (Adaptasi dari Histogram)
sns.countplot(data=dataset, x='outlier_type', ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Jumlah Data per Kategori Outlier', fontsize=14)
axes[0, 0].set_xlabel('Kategori Outlier')
axes[0, 0].set_ylabel('Jumlah')
axes[0, 0].tick_params(axis='x', rotation=15)

# 2. Scatter plot - Sepal Length vs Width
scatter1 = axes[0, 1].scatter(dataset['sepal_length'], dataset['sepal_width'],
                              c=dataset['is_outlier_any'], cmap='coolwarm', alpha=0.8, s=50)
axes[0, 1].set_xlabel('Panjang Sepal')
axes[0, 1].set_ylabel('Lebar Sepal')
axes[0, 1].set_title('Sepal: Panjang vs Lebar', fontsize=14)
fig.colorbar(scatter1, ax=axes[0, 1], label='Status Outlier (1=Ya, 0=Tidak)')

# 3. Scatter plot - Petal Length vs Width
scatter2 = axes[0, 2].scatter(dataset['petal_length'], dataset['petal_width'],
                              c=dataset['is_outlier_any'], cmap='coolwarm', alpha=0.8, s=50)
axes[0, 2].set_xlabel('Panjang Petal')
axes[0, 2].set_ylabel('Lebar Petal')
axes[0, 2].set_title('Petal: Panjang vs Lebar', fontsize=14)
fig.colorbar(scatter2, ax=axes[0, 2], label='Status Outlier (1=Ya, 0=Tidak)')

# 4. Boxplot untuk setiap Fitur berdasarkan Status Outlier
sns.boxplot(data=dataset, x='is_outlier_any', y='petal_length', ax=axes[1, 0])
axes[1, 0].set_title('Boxplot Lebar Petal: Normal vs Outlier', fontsize=14)
axes[1, 0].set_xlabel('Status Outlier (0 = Normal, 1 = Outlier)')
axes[1, 0].set_ylabel('Panjang Petal')

# 5. Visualisasi PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
scatter3 = axes[1, 1].scatter(X_pca[:, 0], X_pca[:, 1], c=dataset['is_outlier_any'],
                              cmap='coolwarm', alpha=0.8, s=50)
axes[1, 1].set_xlabel(f'Komponen Utama 1 ({pca.explained_variance_ratio_[0]:.2%} varians)')
axes[1, 1].set_ylabel(f'Komponen Utama 2 ({pca.explained_variance_ratio_[1]:.2%} varians)')
axes[1, 1].set_title('Visualisasi PCA 2 Komponen', fontsize=14)
fig.colorbar(scatter3, ax=axes[1, 1], label='Status Outlier (1=Ya, 0=Tidak)')

# 6. Heatmap Korelasi Fitur
corr_matrix = dataset[numeric_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=axes[1, 2], fmt='.2f')
axes[1, 2].set_title('Heatmap Korelasi Fitur', fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


# --- ANALISIS TEKSTUAL DARI OUTLIER YANG TERDETEKSI ---
print("\n" + "="*60)
print("Analisis Mendalam dari Outlier yang Terdeteksi")
print("="*60)

# Tampilkan detail dari semua data yang terdeteksi sebagai outlier
outliers_df = dataset[dataset['is_outlier_any'] == 1]
print(f"\nTotal outlier terdeteksi: {len(outliers_df)} data")
print("\nDetail data outlier:")
print(outliers_df[['id', 'class', 'outlier_type'] + numeric_columns])


# Bandingkan statistik antara data normal dan data outlier
normal_data = dataset[dataset['is_outlier_any'] == 0][numeric_columns]
outlier_data = dataset[dataset['is_outlier_any'] == 1][numeric_columns]

print("\n\n" + "="*60)
print("Perbandingan Statistik: Data Normal vs Data Outlier")
print("="*60)
print(f"{'Fitur':<15} {'Rata-rata Normal':<20} {'Rata-rata Outlier':<20} {'Perbedaan':<10}")
print("-" * 70)
for col in numeric_columns:
    normal_mean = normal_data[col].mean()
    outlier_mean = outlier_data[col].mean()
    diff = abs(outlier_mean - normal_mean)
    print(f"{col:<15} {normal_mean:<20.3f} {outlier_mean:<20.3f} {diff:<10.3f}")
```
